## Download Dataset

We will download [the 'best of' cuts from the Watkins Marine Mammal Sound Database](https://cis.whoi.edu/science/B/whalesounds/index.cfm). They have been selected from over 15 000 cuts in total as they were deemed to be of higher sound quality and to contain less noise. You can read more about the origins of this data on [Watkins Marine Mammal Sound Database website](https://cis.whoi.edu/science/B/whalesounds/about.cfm).

Through making this dataset available in this form, we hope to make it easier for this data to be ingested into machine learning pipelines.

```python
import requests
from bs4 import BeautifulSoup, SoupStrainer
import re
```

```python
URL = 'https://cis.whoi.edu/science/B/whalesounds'
page = requests.get(f'{URL}/index.cfm')

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
results = soup.find(class_='visual')
```

```python
name2download_path = {}

for div in results.find_all('div', class_='large-3 columns'):
    name = div.find('h3').text
    name = re.subn("'|\(|\)", '', name)[0]
    download_path = div.find('a')['href']
    name2download_path[name] = download_path
```

```python
%%time

for name, path in name2download_path.items():
    page = requests.get(f'https://cis.whoi.edu/science/B/whalesounds/{path}')
    soup = BeautifulSoup(page.content, 'html.parser')
    dir_name = '_'.join(name.split())
    !mkdir data/{dir_name}
    
    for link in soup.find_all('a'):
        if link.text == 'Download':
            file_to_download = link['href']
            !cd data/{dir_name} && wget --no-check-certificate https://cis.whoi.edu{file_to_download}
```

```python
import glob
import pandas as pd
```

```python
paths = [path[5:] for path in glob.glob('data/**/*.wav')]
species = [path.split('/')[0] for path in paths]
```

```python
anno = pd.DataFrame(data={'path': paths, 'species': species})
```
